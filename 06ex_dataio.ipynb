1\. **Text files**

Perform the following operations on plain `txt` files:

+ create a list of integrer numbers and then save it to a text file named `data_int.txt`. Run the `cat` command to print the content of the file.
+ create a matrix of 5x5 floats and then save it to a text file named `data_float.txt`. Use the `cat` command to print the content of the file.
+ load the `txt` file of the previous point and convert it to a `csv` file by hand.

````python
import pandas as pd
import csv
import numpy as np
import json
````

````python
list_int=[1,2,3,4]
f = open('data/data_int.txt', 'w')
f.write(str(list_int))
f.close()
!cat data/data_int.txt

import numpy as np
matrix = np.arange(1.0,6.0,0.2).reshape(5,5)
with open('data/data_float.txt','w') as f:
    f.write(str(matrix))

print()
!cat data/data_float.txt

with open('data/data_float.txt','r') as file:
    lines = file.readlines()
with open('data/data_float.csv','w', newline='') as output:
    csv_writer = csv.writer(output)
    for line in lines:
        csv_writer.writerow([line.strip()])

!cat data/data_float.csv

    
````

2\. **JSON files**

Load the file `user_data.json`, which can be found at:

- https://www.dropbox.com/s/sz5klcdpckc39hd/user_data.json

and filter the data by the "CreditCardType" when it equals to "American Express". Than save the data to a new CSV file.

````python
data = json.load(open('data/user_data.json'))

data = [person for person in data if person["CreditCardType"] == 'American Express']
print(data)
with open('data/output_csv.csv', 'w', newline='') as output_csv:
    csv_writer = csv.writer(output_csv)

    # Write the header to the CSV file based on the keys of the first dictionary
    csv_writer.writerow(data[0].keys())

    # Write each row to the CSV file
    for person in data:
        csv_writer.writerow(person.values())
````
